```python
# Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca data
df = pd.read_csv('H05.csv')
print(df) # Menampilkan data untuk memudahkan pengerjaan
```

```python
# 1. Menampilkan histogram distribusi nilai matematika
df['nilai_mat'].plot(kind='hist', title='Histogram Distribusi Nilai Matematika', rwidth=0.8)
plt.show()
```

```python
# 2. Menampilkan diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.
df['fakultas'].value_counts().plot(kind='barh', title='Diagram Batang Horizontal Banyaknya Mahasiswa Masing-masing Fakultas')
plt.show()
```

```python
# 3. Menampilkan diagram pie banyaknya mahasiswa masing-masing fakultas.
df['fakultas'].value_counts().plot(kind='pie', title='Diagram Pie Banyaknya Mahasiswa Masing-masing Fakultas')
plt.show()
```

```python
# 4. Menentukan fakultas dengan mahasiswa terbanyak berdasarkan diagram pie dan batang horizontal.
#    Menentukan diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak beserta penjelasannya.
```

Berdasarkan diagram pie, terlihat bahwa FTI dengan potongan pie berwarna biru merupakan potongan terbesar dari fakultas lainnya sehingga fakultas dengan mahasiswa terbanyak adalah FTI. Begitu juga terlihat pada diagram batang horizontal bahwasanya FTI memiliki potongan batang terpanjang (memiliki jumlah mahasiswa terbanyak).

Menurut saya, diagram yang lebih baik untuk menampilkan fakultas dengan mahasiswa terbanyak adalah diagram batang horizontal karena dapat terlihat langsung batang terpanjang serta jumlah mahasiwa untuk tiap fakultas, sedangkan untuk pie chart dapat dilihat bahwasanya FTI dan STEI memiliki ukuran yang hampir sama sehingga jika tidak ditampilkan besar persentasenya, ini justru akan membuat pembaca sulit membedakan manakah fakultas yang memiliki jumlah mahasiswa terbanyak.

```python
# 5. Menampilkan scatter plot dengan nilai kimia sebagai x dan fisika sebagai y.
df.plot(kind='scatter', x='nilai_kim', y='nilai_fis', title='Scatter Plot antara Nilai Kimia dan Fisika')
plt.show()
```
