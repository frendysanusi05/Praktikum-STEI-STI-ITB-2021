```python
# Import library yang dibutuhkan
import pandas as pd
```

```python
# Membaca data
df = pd.read_csv('P05-02.csv')
df # Menampilkan data untuk memudahkan pengerjaan
```

### 1. Banyaknya data

```python
# 1. Menghitung banyaknya data
print('Banyaknya data adalah', len(df), 'data.')
```

### 2. Pajak bangunan per tahun terbesar

```python
# 2. Mencari pajak bangunan per tahun terbesar
terbesar = df['property_tax'].max()

print('Pajak bangunan per tahun terbesar adalah senilai', terbesar)
df.loc[df['property_tax'] == terbesar]
```

### 3. 5 rumah dengan harga sewa per tahun terbesar.

```python
# 3. Mencari 5 rumah dengan harga sewa per tahun terbesar
sewa_terbesar = df.sort_values(['rent_amount'], ascending=[0])
sewa_terbesar[:5]
```

### 4. Rumah yang berada di kota Rio de Janeiro

```python
# 4. Mencari rumah yang berada di kota Rio de Janeiro
rio_de_janeiro = df.loc[df['city'] == 'Rio de Janeiro']
print('Banyak rumah yang berada di kota Rio de Janeiro adalah', len(rio_de_janeiro), 'rumah.')
rio_de_janeiro
```

### 5. Nilai korelasi antara jumlah tempat parkir dengan pajak tanah per tahun. Apakah berkorelasi lurus, terbalik, atau tidak berkolerasi? Jelaskan.

```python
# 5. Menghitung nilai korelasi antara jumlah tempat parkir dengan pajak tanah per tahun
# serta menjelaskan apakah berkorelasi lurus, terbalik, atau tidak berkolerasi
print('Nilai korelasinya adalah', df['parking_spaces'].corr(df['homeowner_tax']))
```

Penjelasan:
Berdasarkan hasil keluaran (output) dari program, kedua atribut jumlah tempat parkir dan pajak tanah per tahun korelasi bernilai positif dan mendekati angka 0. Dengan demikian, dapat disimpulkan bahwa kedua atribut ini tidak berkolerasi antara satu sama lainnya.
