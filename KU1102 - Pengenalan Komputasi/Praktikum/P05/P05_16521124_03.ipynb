```python
# Import library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Membaca data
df = pd.read_csv('P05-03.csv')
df # Menampilkan data untuk memudahkan pengerjaan
```

### 1. Histogram distribusi banyaknya review rumah yang disewakan di wilayah (neighbourhood group) West Region dengan bins 20.

```python
# 1. Menampilkan histogram distribusi banyaknya review rumah yang disewakan di wilayah (neighbourhood group) West Region dengan bins 20.
west_region = df.loc[df['neighbourhood_group'] == 'West Region']
west_region['calculated_host_listings_count'].plot(kind='hist', bins=20, title='Distribusi Banyaknya Review Rumah yang Disewakan di Wilayah West Region')
plt.show()
```

### 2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300.

```python
# 2. Menampilkan pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300.
above300 = df.loc[df['price'] > 300]
above300['neighbourhood'].value_counts().plot(kind='pie', autopct='%1.01f')
plt.plot()
```

### 3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang.

```python
# 3. Menampilkan scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neighbourhood Kallang.
kallang = df.loc[df['neighbourhood'] == 'Kallang']
kallang.plot(kind='scatter', x='longitude', y='latitude')
plt.show()
```

### 4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah dengan 
###    banyak hari ketersediaan rumah dalam satu tahun di atas 100.

```python
# 4. Menampilkan bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah dengan 
#    banyak hari ketersediaan rumah dalam satu tahun di atas 100.
above100 = df.loc[df['availability_365'] > 100]
above100.plot(kind='scatter', x='longitude', y='latitude', sizes=above100['number_of_reviews'])
plt.show()
```

### 5. Stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x dan jenis penyediaan ruang sebagai stacked y.

```python
# 5. Menampilkan stacked bar chart dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x dan jenis penyediaan ruang sebagai stacked y.
richard = df.loc[df['host_name'] == 'Richard']
richard.groupby(["neighbourhood","room_type"])["calculated_host_listings_count"].size().unstack().plot(kind="bar",stacked=True)
plt.show()
```
