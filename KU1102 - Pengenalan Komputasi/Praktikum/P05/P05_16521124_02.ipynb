```python
# Import library yang dibutuhkan
import pandas as pd
```

```python
# Membaca data
df = pd.read_csv('P05-01.csv')
df # Menampilkan data untuk memudahkan pengerjaan
```

### 1. Ada berapa pengguna tidak aktif (Attrited Customer) yang berusia di atas 50 tahun.

```python
# 1. Menghitung banyaknya pengguna tidak aktif (Attrited Customer) yang berusia di atas 50 tahun.
nonaktif_above50 = df.loc[(df['Attrition_Flag'] == 'Attrited Customer') & (df['Customer_Age'] > 50)]
print('Banyaknya pengguna tidak aktif yang berusia di atas 50 tahun adalah', len(nonaktif_above50), 'orang.')
```

### 2. ID pengguna dengan nominal limit kartu kredit terbesar.

```python
# 2. Mencari ID pengguna dengan nominal limit kartu kredit terbesar.
kredit_terbesar = df['Credit_Limit'].max()
print('ID pengguna dengan nominal limit kartu kredit terbesar adalah')
df.loc[df['Credit_Limit'] == kredit_terbesar]
```

### 3. Ada berapa pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40?

```python
# 3. Menghitung banyaknya pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40
blue_above40 = df.loc[(df['Card_Category'] == 'Blue') & (df['Months_on_book'] > 40)]
print('Banyaknya pengguna dengan tipe kartu Blue dan lama penggunaan kartu kredit dalam bulan di atas 40 adalah', len(blue_above40), 'orang.')
```

### 4. Banyak pengguna dengan total nominal yang belum dibayar pengguna kartu kredit sama dengan 0 untuk masing-masing kategori penghasilan.

```python
# 4. Menghitung banyaknya pengguna dengan total nominal yang belum dibayar pengguna kartu kredit sama dengan 0 untuk masing-masing kategori penghasilan.
total0 = df.loc[df['Total_Revolving_Bal'] == 0]
total0['Income_Category'].value_counts()
```

### 5. 5 ID pengguna aktif (Existing Customer) dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.

```python
# 5. Mencari 5 ID pengguna aktif (Existing Customer) dengan banyak bulan kartu kredit tidak aktif dalam 12 bulan terakhir terbanyak.
exist_cust = df.loc[df['Attrition_Flag'] == 'Existing Customer']
terbanyak = exist_cust.sort_values(['Months_Inactive_12_mon'], ascending=[0])
terbanyak[:5]
```
